```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('middle_tn_schools.xlsx')
```

```python
df.head()
```

```python
corr = df.corr()
fig = plt.figure(figsize=(18,8))
ax = sns.heatmap(corr, center = 0, annot=True)
# plt.xtickangle(ax,45)
```

```python
df.isnull().sum()
```

```python
df['state_percentile_15'].fillna(df['state_percentile_15'].mean(),inplace = True)
df['avg_score_15'].fillna(df['avg_score_15'].mean(),inplace = True)
```

```python
fig = plt.figure(figsize = (15,5))
sns.boxplot(data = df)
plt.xticks(rotation = 45)
plt.show()
```

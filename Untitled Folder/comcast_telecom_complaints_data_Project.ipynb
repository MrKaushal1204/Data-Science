```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#  Import data into Python environment.

df = pd.read_csv("Project/Comcast/Comcast_telecom_complaints_data.csv")


# Provide the trend chart for the number of complaints at monthly and daily granularity levels.



# complaint_daily_and _monthly = df[]
df.head(1)
```

```python
df['Month_using_split'] = pd.to_datetime(df['Date_month_year'])

df['Month'] = df['Month_using_split'].dt.month_name()
df['Day'] = df['Month_using_split'].dt.day


# df['month'] = df['Date'].apply( lambda x : int(str(x).split('-')[1]))


# print(df[['Month','Day']].head(3))

groupby_dates = df.groupby('Month')['Customer Complaint'].count()

# print(groupby_dates)

plt.plot(groupby_dates, color = "#046725")
plt.xlabel("Months", size ="15",color ="red")
plt.ylabel("Customer Complain", size ="15", color ="red")
plt.grid(True)
plt.xticks(rotation='vertical')
plt.show()

# print(type(df['Month_using_split']))
   
    
```

```python
df['Customer Complaint'].str.lower().to_frame().value_counts().reset_index().head(10)

```

```python
df['Status'] = df['Status'].apply(lambda x : 'Open' if((x == 'Open') or (x == 'Pending')) else 'Closed')
print(df['Status'].head(10))
```

```python
opn = df[df['Status'] == 'Open'].groupby('State')['Status'].count().to_frame().reset_index()
cls = df[df['Status'] == 'Closed'].groupby('State')['Status'].count().to_frame().reset_index()
```

```python
fig = plt.figure(figsize = (15,15))
plt.barh(cls.State,cls.Status)
plt.barh(opn.State,opn.Status)
plt.xlabel("Status Count", size = '15', color= '#435576')
plt.ylabel("States",size = '15',color ='#435576')
plt.legend(["Closed","Open"])
plt.title("State wise  Status Count", size = '24', color = '#435576')
plt.show()
```
